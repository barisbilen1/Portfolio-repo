```python
                        ##-- Final model & predictions --##
# Importing libraries

import pandas as pd
import getpass
import math
import numpy as np
#from tqdm import trange
import gc
#import matplotlib.pyplot as plt
#import seaborn as sns
#from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix, accuracy_score
from sklearn.metrics import f1_score

if getpass.getuser() == "asus":
    file_path = "C:/Users/asus/Desktop/trendyol_data_v2/"
    
if getpass.getuser() == "your computers name":
    file_path = "C:/Users/..desiredlocation..."
```

```python
data = pd.read_csv(file_path + 'main_data_without_ids.csv') 
```

```python
data2 = data[['Bay','Bayan','discounted_price','ship_cost','coupon_used','is_discounted','promotion_percent',
             'is_shipcost_incurred','shipping_percent','return_prob_from_answers','return_prob_from_reviews',
             'overall_return_prob_product','is_returned']]
```

```python
#Constructing the model.
X = data2.iloc[:, :-1].values
y = data2.iloc[:, -1].values
```

```python
gc.collect()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 41)
```

```python
gc.collect()
```

```python
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
gc.collect()
```

```python
classifier = RandomForestClassifier(n_estimators = 5, criterion = 'entropy', random_state = 41)
classifier.fit(X_train, y_train)
#took 15 minutes to train the model.
```

```python
gc.collect()
```

```python
y_pred_test = classifier.predict(X_test)
print(np.concatenate((y_pred_test.reshape(len(y_pred_test),1), y_test.reshape(len(y_test),1)),1))
```

I don't expect 85% accuracy like before because I had to drop very important \
columns, which possibly constitute a ground for the predictive power \
of the random forest model.

Let's see what the new model gives us.

```python
f1_score(y_test, y_pred_test, average='micro')
```
